The community center claims that the average age of its members is exactly 30 years old.

(Ho): μ = 30  
The average age of members is 30 years

(Ha): μ ≠ 30  
The average age of members is not 30 years

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# 1. Sample Data (Now using a larger sample for Z-test validity)
# Note: Z-tests are more reliable with n > 30
ages = np.array([28, 32, 35, 29, 31, 34, 30, 27, 33, 31, 
                 29, 30, 32, 28, 30, 31, 34, 29, 30, 32,
                 33, 27, 28, 31, 30, 32, 29, 30, 34, 31])

# 2. Set the Population Mean to test against (Null Hypothesis)
pop_mean_null = 30

# 3. Perform One-Sample Z-Test
# value: the mean under the null hypothesis
z_stat, p_val = ztest(ages, value=pop_mean_null)

# 4. Print Results
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_val:.4f}")

# 5. Decision Logic
alpha = 0.05
if p_val < alpha:
    print(f"Decision: Reject H0 (p < {alpha})")
else:
   print(f"Decision: Fail to Reject H0 (p >= {alpha})") 
```

```python
ages_outlier = np.append(ages, 110)

# Run Z-test again
z_stat2, p_val2 = ztest(ages_outlier, value=pop_mean_null)

print(f"New Z-Statistic: {z_stat2:.4f}")
print(f"New P-Value: {p_val2:.4f}")

if p_val2 < alpha:
    print(f"Decision: Reject H0 (p < {alpha})")
else:
    print(f"Decision: Fail to Reject H0 (p >= {alpha})")
```
